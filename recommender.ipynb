#### This genres are in weird fromate and we need to fetch only important tags like, Action, Adventure, etc. But the thing here is it is in the string formate, so we will use 'ast.literal_eval' helper function to convert these sting formates into dictionaries and then we will be able to fetch the particular name tag form those strings

```python
# We need formate like ['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
)
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

#### We will do the same for keywords

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

##### We have to do the same thing for cast but for the first few actors

```python
movies['cast'][1]
```

```python
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 4:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
        
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast'][1]
```

```python
movies['crew'][0]
```

#### We only need the value where job is director

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

#### We are having words like, (adult,adults), (activity,activities) so we will apply stemming to remove all unnecessary plurals from our tags

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction culture clash future space war space colony society space travel futuristic romance space alien tribe alien planet cgi marine soldier battle love affair anti war power relations mind and soul 3d samworthington zoesaldana sigourneyweaver stephenlang jamescameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
